```python
import os
import imageio
import numpy as np
import matplotlib.pyplot as plt
import shutil
import ray
from ray.rllib.env.wrappers.pettingzoo_env import ParallelPettingZooEnv
from ray.tune.registry import register_env
from pettingzoo.sisl import waterworld_v4
from ray.rllib.algorithms.ppo import PPOConfig



print(f"Ray: {ray.__version__}")
try:
    import pettingzoo
    print(f"PettingZoo: {pettingzoo.__version__}")
except ImportError:
    print("PettingZoo: Not Found")
print(f"NumPy: {np.__version__}")
try:
    import gymnasium
    print(f"Gymnasium: {gymnasium.__version__}")
except ImportError:
    print("Gymnasium: Not Found")

if ray.is_initialized():
    ray.shutdown()
ray.init(ignore_reinit_error=True, num_cpus=2, num_gpus=0)

def env_creator(env_config=None):
    env = waterworld_v4.parallel_env(max_cycles=500)
    return env

try:
    test_env = env_creator()
    test_env.observation_space(test_env.possible_agents[0])
    test_env.reset()
    test_env.close()
except Exception as e:
    print(f"ERROR creating test_env: {e}")
    raise

ENV_NAME = "waterworld_vector_v1"
register_env(ENV_NAME, lambda config_dict: ParallelPettingZooEnv(env_creator(config_dict)))

dummy_rllib_env = ParallelPettingZooEnv(env_creator())
agents = list(dummy_rllib_env.observation_space.keys())
if not agents:
     raise ValueError("Could not determine agents.")

ppo_config = (
    PPOConfig()
    .environment(env=ENV_NAME, clip_rewards=True)
    .framework("torch")
    .rollouts(num_rollout_workers=1)
    .training(model={"fcnet_hiddens": [128, 128]})
    .multi_agent(
        policies={agent: (None, dummy_rllib_env.observation_space[agent], dummy_rllib_env.action_space[agent], {}) for agent in agents},
        policy_mapping_fn=(lambda agent_id, episode, worker, **kwargs: agent_id),
    )
    .resources(num_gpus=0)
    .debugging(log_level="WARN")
)

algo = ppo_config.build()

NUM_TRAINING_ITERATIONS = 100
episode_rewards_history = []

print(f"Starting training for {NUM_TRAINING_ITERATIONS} iterations...")
for i in range(NUM_TRAINING_ITERATIONS):
    result = algo.train()
    reward_mean = result['episode_reward_mean']
    episode_rewards_history.append(reward_mean)
    if (i + 1) % 10 == 0:
        print(f"Iter. {i + 1:3d}: reward_mean = {reward_mean:.2f}, ep_len_mean = {result.get('episode_len_mean', 0):.2f}")
print("Training completed.")

plt.rcParams.update({
    'figure.facecolor': '#2b2b2b',     # fondo exterior
    'axes.facecolor':   '#2b2b2b',     # fondo del área del gráfico
    'axes.edgecolor':   '#cccccc',     # borde de los ejes
    'axes.labelcolor':  '#cccccc',     # etiquetas de los ejes
    'xtick.color':      '#cccccc',     # color de los números en el eje X
    'ytick.color':      '#cccccc',     # color de los números en el eje Y
    'text.color':       '#cccccc',     # color del título y textos
    'grid.color':       '#444444',     # color de la grilla
})

plt.figure(figsize=(12, 6))
plt.plot(episode_rewards_history)
plt.title(f"Episode Reward Mean ({ENV_NAME})")
plt.xlabel("Training Iteration")
plt.ylabel("Episode Reward Mean")
plt.grid(True)
plt.savefig("/content/reward_plot.png")
plt.show()

print("Creating evaluation video...")
eval_env_render = waterworld_v4.parallel_env(render_mode="rgb_array", max_cycles=300)
observations, infos = eval_env_render.reset()
frames = []
NUM_VIDEO_FRAMES = 200

for step in range(NUM_VIDEO_FRAMES):
    actions_to_take = {}
    current_active_agents = list(eval_env_render.agents)
    if not current_active_agents:
        break
    for agent_id in current_active_agents:
        if agent_id in observations:
            action_tuple_or_action = algo.compute_single_action(
                observation=observations[agent_id], policy_id=agent_id, explore=False
            )
            actions_to_take[agent_id] = action_tuple_or_action[0] if isinstance(action_tuple_or_action, tuple) else action_tuple_or_action
    if not actions_to_take:
        break
    next_observations, rewards, dones, truncations, infos = eval_env_render.step(actions_to_take)
    observations = next_observations
    try:
        frame = eval_env_render.render()
        frames.append(frame)
    except Exception as e:
        print(f"Error during render(): {e}. Skipping frame.")
        break
    all_terminated = True
    if current_active_agents:
        all_terminated = all(dones[agent_id] or truncations[agent_id] for agent_id in current_active_agents)
    if all_terminated:
        break
eval_env_render.close()

VIDEO_FILENAME = "waterworld_eval_video.mp4"
output_path_tmp = f"/tmp/{VIDEO_FILENAME}"
output_path_content = f"/content/{VIDEO_FILENAME}"

if frames:
    imageio.mimsave(output_path_tmp, frames, fps=15, macro_block_size=1)
    if os.path.exists(output_path_tmp):
        shutil.move(output_path_tmp, output_path_content)
        print(f"Video saved to: {output_path_content}")
else:
    print("No frames generated for video.")

ray.shutdown()
print("--- Script finished ---")
```
